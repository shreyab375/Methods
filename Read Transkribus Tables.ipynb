## Reading the Transkribus data for 1 year

```python
#import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Give the entire dataframe of collections and the year as input
def read_data(df, year):
    df.rename(columns={'Unnamed: 1':'Day'},inplace=True)
    df = df[df.Day.notna()]
    dd = dict(tuple(df.groupby('p001.jpg')))
    n =  np.array(list(dd.keys()))
    dd1 = {}
    temp = pd.DataFrame()
    data = pd.DataFrame()
    i=0
    Year = "1949"
    Month = ["01", "02","03","04","05","06","07","08","09","10","11","12"]
    for i in range(len(n)):
        data = pd.DataFrame()
        temp = pd.DataFrame()
        m=0
        Day = dd[n[i]].Day
        b = pd.DataFrame(dd[n[i]])
        #print(b)
        #print(Day)
        for j in range(0,24,2):
            temp["Day"] = b.Day
            temp["Month"] = Month[m]
            temp["Year"] = Year
            temp["G+S"] = b[b.columns[j+2:j+3]]
            temp["k"] = b[b.columns[j+3:j+4]]
            m= m+1
            data = data.append(temp)
            
        dd1[n[i]] = data
        
        
        
    return dd1
```

```python
#Read the table for one year
df = pd.read_csv(r'C:\Users\ShreyaB\Documents\Mitteilungen\\data\Transkribus\Tables1949_pdf_tables.csv')
```

```python
#call the function
dd1 = read_data(df, 1949) 
```
